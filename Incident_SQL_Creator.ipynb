```python
from numpy import genfromtxt
from time import time
from datetime import datetime
from sqlalchemy import Column, Integer, Float, Date, String, Time, DateTime
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import create_engine
from sqlalchemy.orm import sessionmaker
import csv
from dateutil import parser
```

```python
def Load_Data(file_name):
#     data = genfromtxt(file_name, delimiter=',', skip_header=1, converters={0: lambda s: str(s)})
#     return data.tolist()
    with open(file_name, "r") as f:
        dr = csv.DictReader(f)
        rows = []
        for i in dr:
            rows.append(i)
        
        return rows
```

```python
Base = declarative_base()
```

```python
class Incident(Base):
    __tablename__ = 'Incident_History'
    __table_args__={'sqlite_autoincrement': True}
    id = Column(Integer, primary_key=True, nullable=False)
    direction = Column(String)
    color = Column(String)
    location = Column(String)
    problem = Column(String)
    delay = Column(Float)
    full_text= Column(String)
    time = Column(Time)
    date = Column(Date)
```

```python
if __name__ == "__main__":
    engine = create_engine('sqlite:///incident.sqlite')
    Base.metadata.create_all(engine)
    
    session = sessionmaker()
    session.configure(bind=engine)
    s = session()
    
    file_name = 'incident.csv'
    data = Load_Data(file_name)
    print(len(data))
  

    for i in data:
#         print(i)
        record = Incident(**{
            'direction': i['direction'],
            'color':i['color'],
            'location': i['location'],
            'problem': i['problem'],
            'delay': i['delay (min)'],
            'full_text': i['full text'],
            'time': parser.parse(i['time']).time(),
            'date': parser.parse(i['date'])
            })

#         record = Incident(**{
#             'direction': i[0],
#             'color':i[1],
#             'location': i[2],
#             'problem': i[3],
#             'delay': i[4],
#             'full_text': i[5],
#             'time': i[6],
#             'date': i[7]
#             })
        s.add(record)
    s.commit()
```
