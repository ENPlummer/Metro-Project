```python
import csv
import pandas as pd
import numpy as np
import os
import sqlalchemy
import sqlite3
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect , desc, distinct, func, extract, and_
from odo import odo, resource, discover
from flask import Flask, jsonify, render_template, send_from_directory
```

```python
incident_data = pd.read_csv('Data/Metro-Incident-Data.csv')
incident_data.head()
```

```python
new_time = pd.to_datetime(incident_data['time'])
```

```python
new_df = incident_data.drop(['time'], axis =1)
new_df = pd.concat([new_df, new_time], axis =1)
new_df['date'] = new_df['time'].dt.date
new_df['time'] = new_df['time'].dt.time
new_df = new_df.dropna(how='any')
new_df
```

```python
new_df.to_csv('incident.csv', encoding='utf-8', index = False)
```
