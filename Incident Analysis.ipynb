```python
import csv
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
incident_data = pd.read_csv('Metro-Incident-Data.csv')
incident_data.head()
```

```python
new_time = pd.to_datetime(incident_data['time'])
```

```python
new_df = incident_data.drop(['time'], axis =1)
new_df = pd.concat([new_df, new_time], axis =1)
new_df.head()
```

```python
disk_engine = create_engine('sqlite:///incident_db.db')
new_df.to_sql('incident_db',disk_engine)
```

```python
incident_db = pd.read_sql_query('SELECT * FROM incident_db',disk_engine)
incident_db.head()
```
